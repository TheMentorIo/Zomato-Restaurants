```python
countries['Country'].loc[countries['Country'] == 'India']= "IN"
countries['Country'].loc[countries['Country'] == 'Australia']= "AU"
countries['Country'].loc[countries['Country'] == 'South Africa']= "SA"
countries['Country'].loc[countries['Country'] == 'Brazil']= "BR"
countries['Country'].loc[countries['Country'] == 'Canada']= "CA"
countries['Country'].loc[countries['Country'] == 'Indonesia']= "ID"
countries['Country'].loc[countries['Country'] == 'New Zealand']= "NZ"
countries['Country'].loc[countries['Country'] == 'Phillipines']= "PHL"
countries['Country'].loc[countries['Country'] == 'Qatar']= "QA"
countries['Country'].loc[countries['Country'] == 'Singapore']= "SG"
countries['Country'].loc[countries['Country'] == 'South Africa']= "SA"
countries['Country'].loc[countries['Country'] == 'Sri Lanka']= "SL"
countries['Country'].loc[countries['Country'] == 'Turkey']= "TR"
countries['Country'].loc[countries['Country'] == 'United Kingdom']= "UK"
countries['Country'].loc[countries['Country'] == 'United States']= "US"
countries
```

```python
countries.set_index('Country Code', inplace=True)
zomato.set_index('Country Code', inplace=True)
zomato=zomato.join(countries,how="inner").reset_index()
countries=countries.reset_index()
zomato.drop('Country Code',axis=1,inplace=True)
```

```python
def league(row):
    x=row.split("(")
    if x[0]== "Dollar":
        return "US$"
    elif x[0] == "Pounds":
        return "£"
    elif x[0] == "NewZealand":
        return "NZ$"
    else:
        return x[1].split(")")[0]
zomato["Currency"]=zomato["Currency"].apply(league)
zomato["Currency"].unique()
```

```python
df = zomato.assign(Cuisines=zomato['Cuisines'].str.split(',')).explode('Cuisines')
df['Cuisines'] = df['Cuisines'].str.strip()
df_grouped = df.groupby(["Country", "Cuisines"])["City"].count().reset_index()

df_grouped_sorted = df_grouped.sort_values(by=["Country", "City"], ascending=[True, False])
top_5_cuisines = df_grouped_sorted.groupby('Country').head(5)
unique_countries = top_5_cuisines['Country'].unique()

# Calculate number of columns needed based on the number of unique countries
num_cols = (len(unique_countries) + 2) // 3  # Ensure at least 3 rows

# Create subplots with specified layout
fig, axs = plt.subplots(nrows=3, ncols=num_cols, figsize=(15, 10))

# Flatten the axs array to iterate over all subplots
axs = axs.flatten()

for i, country in enumerate(unique_countries):
    data = top_5_cuisines[top_5_cuisines['Country'] == country]
    ax = axs[i]
    ax.pie(data['City'], labels=data['Cuisines'], autopct='%1.1f%%', startangle=140)
    ax.set_title(f"Top 5 Cuisines Density in {country}")
    ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Hide any remaining empty subplots
for j in range(i + 1, len(axs)):
    axs[j].axis('off')

plt.savefig("Top 5 Cuisines Density in each Country")
```

```python
df = zomato.assign(Cuisines=zomato['Cuisines'].str.split(',')).explode('Cuisines')
df=df.groupby(["Cuisines","Rating text"])["City"].count().reset_index()

fig, axes = plt.subplots(1, 3, figsize=(25, 8), sharey=True)

# Iterate over each rating text and create a subplot
for i, rating_text in enumerate(['Excellent', 'Good','Average']):
    ax = axes[i]
    data_subset = df[df["Rating text"] == rating_text].sort_values("City", ascending=False)[:5]
    sns.barplot(data=data_subset, x="Cuisines", y="City", ax=ax)
    ax.set_yscale("log")
    ax.set_ylabel("Density", fontsize=20)
    ax.set_xlabel("Foods", fontsize=20)
    ax.set_title(f"Top 5 {rating_text} Foods Density upon all Country", fontsize=20)
    ax.tick_params(axis='x', labelsize=15)
    ax.tick_params(axis='y', labelsize=15)

plt.savefig('plot_top_5_foods_ratings.png')
```
